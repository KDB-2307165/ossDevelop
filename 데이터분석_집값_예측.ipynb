## 새로운 데이터 수정

```python
# prompt: /content/sample_data/california_housing_train.csv를 분석해주세요.

import pandas as pd

# Load the dataset
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Display basic information about the dataset
print("First 5 rows of the dataset:")
print(df.head())
print("\nData types of each column:")
print(df.dtypes)
print("\nSummary statistics of numerical columns:")
print(df.describe())
print("\nNumber of missing values in each column:")
print(df.isnull().sum())
print("\nCorrelation matrix:")
print(df.corr())

# Example of more specific analysis (you can customize this part)
# Calculate the average median house value for each housing median age category
# print("\nAverage median house value for each housing median age category:")
# print(df.groupby('housing_median_age')['median_house_value'].mean())

# You can add more analysis based on your needs.
# For example, you can visualize the data using libraries like matplotlib or seaborn.
```

```python
# prompt: 차트도 만들어주세요.

import matplotlib.pyplot as plt
import seaborn as sns

# Example 1: Scatter plot of median house value vs. median income
plt.figure(figsize=(10, 6))
sns.scatterplot(x='median_income', y='median_house_value', data=df)
plt.title('Median House Value vs. Median Income')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()

# Example 2: Histogram of housing median age
plt.figure(figsize=(8, 5))
sns.histplot(df['housing_median_age'], bins=20, kde=True)
plt.title('Distribution of Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Frequency')
plt.show()

# Example 3: Box plot of median house value for different ocean proximity
plt.figure(figsize=(12, 6))
sns.boxplot(x='ocean_proximity', y='median_house_value', data=df)
plt.title('Median House Value by Ocean Proximity')
plt.xlabel('Ocean Proximity')
plt.ylabel('Median House Value')
plt.xticks(rotation=45)
plt.show()

# Example 4: Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
